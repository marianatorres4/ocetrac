```python
import xarray as xr
import numpy as np
import pandas as pd

def convert_dataframe_to_dataset(df):
    ds = xr.Dataset()

    # Handle scalar columns
    for col in ['id', 'duration', 'total_area', 'intensity_cumulative']:
        ds[col] = ('mhw', df[col].values)

    # Directly add the object columns without converting to numpy arrays
    ds['dates'] = xr.DataArray(df['dates'].values, dims='mhw')
    ds['coords'] = xr.DataArray(df['coords'].values, dims='mhw')
    ds['centroid'] = xr.DataArray(df['centroid'].values, dims='mhw')
    ds['intensity_max'] = xr.DataArray(df['intensity_max'].values, dims='mhw')
    ds['intensity_mean'] = xr.DataArray(df['intensity_mean'].values, dims='mhw')
    ds['intensity_min'] = xr.DataArray(df['intensity_min'].values, dims='mhw')

    return ds

# Assume `df` is the DataFrame created by `collect_surface_stats`
df = tracker.collect_surface_stats(labels)

# Convert the DataFrame to an xarray.Dataset
ds = convert_dataframe_to_dataset(df)

# Save the dataset to a NetCDF file
#ds.to_netcdf('events.nc')
```
